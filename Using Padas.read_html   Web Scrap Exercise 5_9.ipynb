## Basketball Page

```python
year="2020"
urlink="https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"
```

```python
#{} is used to replace the {}
```

```python
url=urlink.format(year)
```

```python
import pandas as pd
```

```python
df1=pd.read_html(url,header=0)
```

```python
df1   # this is not in the dataframe format
```

```python
len(df1)  ## If the webpage has multiple table, 
          ##there will be many tables. This one shows 1, so there is only 1 table in the web
```

```python
df1[0]   # it is dataframe format
```

```python
Basketball=df1[0]
```

```python
Basketball
```

```python
Basketball[Basketball.Age=="Age"]  ## Remove those rows because the website has the special design
```

```python
Basketball=Basketball.drop(Basketball[Basketball.Age=="Age"].index)
```

```python
import seaborn as sns
```

```python
sns.distplot(Basketball.PTS,kde=False)
```

```python
sns.distplot(Basketball.PTS,kde=False,hist_kws=dict(edgecolor="Green",linewidth=2),color="#00BFc5")
```
